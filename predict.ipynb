```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pandas as pd
import numpy as np
import pickle
```

```python
model=load_model('model.h5')
with open('onehot.pkl','rb') as f:
    one_hot=pickle.load(f)
with open('label.pkl','rb') as f:
    label=pickle.load(f)
with open('scaler.pkl','rb') as f:
    scaler=pickle.load(f)
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
geo_encoded=one_hot.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=one_hot.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_df=pd.DataFrame([input_data])
input_df
```

```python
import pandas as pd
input_df=pd.concat([input_df.drop("Geography",axis=1),geo_encoded_df],axis=1)
input_df
```

```python
input_df['Gender']=label.transform(input_df['Gender'])
```

```python
## Scaling the input data
input_scaled=scaler.transform(input_df)
input_scaled
```

```python
## PRedict churn
prediction=model.predict(input_scaled)
prediction
```

```python
prediction_proba = prediction[0][0]
```

```python
prediction_proba
```

```python
if prediction_proba > 0.5:
    print('The customer is likely to churn.')
else:
    print('The customer is not likely to churn.')
```
